```python
import numpy as np
x = np.arange(100)
y = [i**2 for i in x]
y = np.array(y)
```

```python
x.shape = (100,1)
```

```python
x
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(x, y)

y_predictions = model.predict(x)



```

```python
y_predictions
```

```python
import matplotlib.pyplot as plt

plt.plot(x, y, c='red', label='actual data', marker='o')
plt.plot(x, y_predictions, c='green', label ='predictions', marker='*')
plt.xlabel('X')
plt.ylabel('y')
plt.title('linear regression')

plt.legend()
```

```python
from sklearn.metrics import mean_squared_error


print("mean squared error is ", mean_squared_error(y, y_predictions))
```

```python
model.coef_[0]
```

```python
import yfinance as yf

reliance_df = yf.download('RELIANCE.NS', start = "2020-10-01", end ="2024-07-18")
 
```

```python
reliance_df.info()
```

```python
reliance_df.tail()
```

```python
reliance_df = reliance_df[:-1]
```

```python
reliance_df.shape
```

```python
reliance_df['Tomorrow Close'] = reliance_df['Close'].shift(-1)
```

```python
x = reliance_df.drop('Tomorrow Close', axis =1)
```

```python
y = reliance_df['Tomorrow Close']
```

```python
x.shape[0]
```

```python
x_train = x[:int(x.shape[0]*0.8)]
x_test = x[int(x.shape[0]*0.8):]

y_train = y[:int(y.shape[0]*0.8)]
y_test = y[int(y.shape[0]*0.8):]
```

```python
x_test.shape
```

```python
from sklearn.linear_model import LinearRegression
lr_reg = LinearRegression()

lr_reg.fit(x_train, y_train)

y_predictions = lr_reg.predict(x_test)
```

```python
y_predictions.shape
```

```python
x
```

```python
import matplotlib.pyplot as plt

plt.scatter(x_test.index, y_test, c='green', label='actual tommorow close price')
plt.scatter(x_test.index, y_predictions, c ='red', label = 'lr predictions')
plt.xlabel('Date')
plt.ylabel('tomorow close price')
plt.title('acutal close prices vs linear regression predictions')
```

```python
from sklearn.metrics import mean_squared_error

print('mean square error of lr on reliance stock data is ', mean_squared_error(y_test, y_predictions))
```
