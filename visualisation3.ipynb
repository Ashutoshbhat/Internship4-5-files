```python
import matplotlib.pyplot as plt

list1 =[1,2,3,4,5,6]
list2 =[2,4,9,16,25,36]

plt.plot(list1, list2, marker='o', linestyle='--', color='red', markersize=10,label='list1 vs list2')
plt.plot(list1, list1, marker='*',color='green', label ='list1 vs list 1')
plt.legend()
```

```python
plt.scatter(list1, list2)
plt.title('scatter plot of list1 and list2')
plt.xlabel('list 1')
plt.ylabel('list 2')
```

```python
import pandas as pd

titanic_df = pd.read_csv('titanic.csv')
```

```python
titanic_df.columns
```

```python
titanic_df.dropna(subset=['Age','Fare'], inplace =True)
```

```python
titanic_df.sort_values(by='Age', inplace=True)
```

```python
titanic_df
```

```python
plt.plot(titanic_df['Age'], titanic_df['Fare'])
plt.title('age vs fare')
plt.xlabel('AGe')
plt.ylabel('FAre')
```

```python
titanic_df.boxplot(column='Age', by='Pclass')
plt.suptitle('AGe distribution based on sex')

```

```python
titanic_df.columns
```

```python
import seaborn as sns

sns.pairplot(titanic_df[['Age', 'Fare', 'Survived', 'Pclass']], hue='Survived')
```

```python
sns.barplot(x='Pclass',y='Fare', data=titanic_df)
```

```python
sns.kdeplot(data=titanic_df, x='Age', hue='Sex',fill=True)
```

```python
plt.hist(titanic_df['Age'], bins=20,edgecolor ='r')
```

```python
survival_counts = titanic_df['Survived'].value_counts()
```

```python
survival_counts.plot(kind='bar')
```

```python
survival_counts.plot(kind='pie',labels=['Survived','not survived'], autopct='%1.1f%%')
```

```python
fig, axis =plt.subplots(2,1)
titanic_df['Age'].plot(kind='hist', ax= axis[0],  color ='red', edgecolor='k')
axis[0].set_title('age histogram')

titanic_df['Fare'].plot(kind='hist', ax =axis[1], color ='green', edgecolor='r')
axis[1].set_title('fare histogram')
```

```python
titanic_df.columns
```

```python
cor_matrix = titanic_df[['Age', 'Fare','Pclass','Survived']].dropna().corr()
```

```python
cor_matrix
```

```python
sns.heatmap(cor_matrix, annot=True, linewidths=3, cmap='coolwarm')
```

```python
sns.violinplot(x='Sex', y = 'Age', data =titanic_df)
plt.title('age distribtuion w.r.t sex using violin plot')
```
