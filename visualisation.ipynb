```python
list_salary =[1000, 1200, 1300, 1400, 1500]

list_savings =[300, 400, 500, 600, 700]
```

```python
plt.plot(list_salary, list_savings,c='red', linestyle ='--', marker = 'o', label = 'savings vs income')
plt.plot(list_salary, list_salary, label ='salary vs salary', marker ='*', c='blue')

plt.xlabel('employee salary')
plt.ylabel('employee savings')
plt.legend()

```
