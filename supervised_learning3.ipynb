```python
import numpy as np
x = np.arange(100)
y = [i**2 for i in x]
y = np.array(y)
```

```python
x.shape = (100,1)
```

```python
x
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(x, y)

y_predictions = model.predict(x)



```

```python
y_predictions
```

```python
import matplotlib.pyplot as plt

plt.plot(x, y, c='red', label='actual data', marker='o')
plt.plot(x, y_predictions, c='green', label ='predictions', marker='*')
plt.xlabel('X')
plt.ylabel('y')
plt.title('linear regression')

plt.legend()
```

```python
from sklearn.metrics import mean_squared_error


print("mean squared error is ", mean_squared_error(y, y_predictions))
```
