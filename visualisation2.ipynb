```python
titanic_df.dropna(inplace=True)
```

```python
titanic_df['Embarked'].unique()
```

```python
dict_embarked_encoding = {'C':0, 'S':1, 'Q':2}
```

```python
titanic_df['Cabin'].unique
```

```python
titanic_df['Embarked'] = titanic_df['Embarked'].map(dict_embarked_encoding)
```

```python
cor_matrix= titanic_df[['Age','Fare','Embarked']].dropna().corr()
```

```python
import seaborn as sns
plt.plot(figsize=(10,8))
sns.heatmap(cor_matrix, annot=True)
```
